```
Algoritmo contadorDigitos
	Leer Num
	Total <- 0
	Mientras Num>0 Hacer
		Num <- trunc(Num/10)
		Total <- Total+1
	FinMientras
	Escribir Total
FinAlgoritmo
```



###Codigo de Python


```
"""
Nombre: Ejercicio1
Entrada: Valor de 4 parametros
Salida: resultado de la suma de cuatro numeros divido entre 4
Restricciones: deber ser solo numeros enteros
"""
def ejercicio1(Num1, Num2, Num3, Num4):
    if isinstance(Num1, int) and isinstance(Num2, int) and isinstance(Num3, int) and isinstance(Num4, int):
        result = (Num1 + Num2 + Num3 + Num4) / 4

    return result
    
```

```
"""
Nombre: Ejercicio7
Entrada: solo un parametro
Salida: Calcular el descuento del traje dependiendo en su precio y su suma total
Restricciones: 
"""
def ejercicio7(Num):
    if isinstance(Num, int):
        if Num > 2500:
            Descuento15 = Num * 0.15
            print(Descuento15)
            total = Num - Descuento15
            print(total)

        else:  
            Descuento8 = Num * 0.08
            print(Descuento8)
            total = Num - Descuento8
            print(total)
```
