###Pseudocodigo


```
Algoritmo Ejercicio1
	Leer N1
	Leer N2
	Leer N3
	Leer N4
	Total = (N1 + N2 + N3 + N4)/ 4
	Escribir Total
FinAlgoritmo

```



###Codigo de Python


```
"""
Nombre: Ejercicio1
Entrada: Valor de 4 parametros
Salida: resultado de la suma de cuatro numeros divido entre 4
Restricciones: deber ser solo numeros enteros
"""
def ejercicio1(Num1, Num2, Num3, Num4):
    if isinstance(Num1, int) and isinstance(Num2, int) and isinstance(Num3, int) and isinstance(Num4, int):
        result = (Num1 + Num2 + Num3 + Num4) / 4

    return result
    
```

###Pseudocodigo


```
Algoritmo Ejercicio7
	Leer Traje
	Si Traje>2500 Entonces
		Descuento15 <- Traje * 0.15
		Escribir Descuento15
		Total15 <- Traje - Descuento15
		Escribir Total15
	SiNo
		Descuento8 <- Traje * 0.08
		Escribir Descuento8
		Total8 <- Traje - Descuento8
		Escribir Total8
	FinSi
FinAlgoritmo
```



###Codigo de Python


```
"""
Nombre: Ejercicio7
Entrada: solo un parametro
Salida: Calcular el descuento del traje dependiendo en su precio y su suma total
Restricciones: 
"""
def ejercicio7(Num):
    if isinstance(Num, int):
        if Num > 2500:
            Descuento15 = Num * 0.15
            print(Descuento15)
            total = Num - Descuento15
            print(total)

        else:  
            Descuento8 = Num * 0.08
            print(Descuento8)
            total = Num - Descuento8
            print(total)
```
